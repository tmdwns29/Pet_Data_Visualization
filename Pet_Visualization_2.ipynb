```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import koreanize_matplotlib

import folium
```

```python
df = pd.read_csv('/content/drive/MyDrive/ABC부트캠프/2022동물복지에 대한 국민인식 조사.csv', encoding='UTF-8')
df.info()
```

```python
plt.figure(figsize=(10, 5))

plt.title('동물 복지에 대한 국민인식 조사')
ax=sns.barplot(data=df, y=df['이유'], x=df['비율'], palette='Oranges_r', width=0.5)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
# 전국 반려동물 시설 비율
plt.figure(figsize=(7,7))

sizes = df['이유'].value_counts()
explode = [0.01] * len(sizes)

wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}
colors = ['#FF5722', '#FB8C00', '#FB8C00', '#FB8C00', '#FB8C00','#FB8C00','#FB8C00']

plt.pie(df['비율'], labels=df['이유'], autopct='%.2f%%', explode=explode, wedgeprops=wedgeprops, colors=colors)
plt.title('동물 복지에 대한 국민인식 조사')
plt.show()
```
